### Question 1 

Vu la forme des équations de Maxwell données dans l'énoncé, la permittivité $\mu$ dans le milieu considéré est donc la permittivité du vide $\mu_0$. 

Prenons l'équation $$ \vec{\nabla} \times \vec{B}  = \mu_0 \vec{J}$$
On sait que $ \vec{J} = \sigma \vec{E} $, donc $$  \vec{\nabla} \times \vec{B}  = \mu_0 \sigma \vec{E}$$ 
En prenant le rotationnel du champ $\vec{E}$, on obtient alors 

$$\vec{\nabla} \times \vec{E} = \frac{1}{\mu_0 \sigma} \vec{\nabla} \times \vec{\nabla} \times \vec{B} = \frac{1}{\mu_0 \sigma} (\vec{\nabla}(\vec{\nabla}.\vec{B}) - \Delta \vec{B}) = - \partial_t \vec{B} $$ 

où la dernière égalité est obtenue par la première équation de Maxwell. La dernière équation de Maxwell utile dans notre problème nous disant que la divergence de $\vec{B}$ est nulle nous permet d'obtenir l'équation de diffusion de ce champ magnétique : 

$$ \frac{\partial\vec{B}}{\partial t} - \frac{1}{\mu_0 \sigma} \nabla^2\vec{B} = 0$$


ce qui correspond bien à une équation de diffusion dont le coefficient de diffusion, appelons-le $\alpha$, est $\alpha = \frac{1}{\mu_0 \sigma}$. On voit que c'est plausible car, dans le cas limite d'un supraconducteur (i.e. $\sigma$ est infini), le coefficient de diffusion est nul ce qui est logique puisque le champ magnétique y est nul et ne se diffuse donc pas du tout.

Par symétrie, le problème est indépendant de $z$ puisque le conducteur est infini. Ainsi, $\vec{B} = \vec{B}(x,y;t)$. De plus, on peut voir le courant appliqué à la surface extérieur comme formant une boucle de courant (et, en considérant $z$, comme un solénoïde), ce qui nous renseigne sur la direction de $\vec{B}$ tout simplement par la "règle de la main droite": celui ci sera dirigé exclusivement dans la direction $z$ et plus précisément selon $-\vec{u_z}$.

Finalement, on a donc $\vec{B} = B(x,y;t) \, \vec{u_z} =B_z \, \vec{u_z} $ avec $B \leq 0$, ceci $\forall t \geq 0 $ et $ \forall (x,y)$ dans le domaine. De plus, il évolue selon : 

$$\frac{\partial B_z}{\partial t} - \frac{1}{\mu_0 \sigma} \nabla^2 B_z = 0$$

*Remarque*

Dans ce travail, nous noterons $B$ ou $B_z$ sans distinction, il s'agit bien de la seule composante non-nulle du champ magnétique.

### Question 2

Nous allons établir les conditions aux bords du domaine considéré.

##### Bords du conducteur en contact avec le vide

Commençons par les bords extérieurs. Pour déterminer la valeur du champ magnétique sur les bords extérieurs du conducteur, on peut utiliser le théorème d'Ampère, donné par:

$$ \oint_{\partial S} \vec{B}.\vec{dl}=\mu_0~I_{traversant} $$
où l'intégrale est une intégrale le long du bord $\partial S$, où $\vec{dl}$ est le vecteur infinitésimal tangent au bord de la surface S considérée et où $I_{traversant}$ est le courant total traversant la surface S considérée.

Essayons de trouver une condition pour le bord gauche. Pour ce faire, on choisit de considérer une surface rectangulaire perpendiculaire au courant $\vec{J}$ du bord gauche, ayant des côtés de longueurs $\Delta z$ selon la direction z et $\Delta x$ selon la direction x. On place ce rectangle de manière à ce que le courant $\vec{J}$ le traverse en son milieu. On doit également choisir le sens des vecteurs tangents à cette surface, on considère que pour le bord du rectangle selon la direction z le plus à gauche, le vecteur tangent est dirigé selon $-\vec{u_z}$. Les autres vecteurs tangents suivent alors ce sens. Remarquons que le vecteur tangent reste constant sur un côté du rectangle.

En effectuant l'intégrale du membre de gauche, on obtient:

$$\oint_{\partial S} \vec{B}.\vec{dl}=\int_{bord~ext} \vec{B_{ext}}.\vec{dl} + \int_{bord~int} \vec{B_{int}}.\vec{dl}=-\int_{bord~ext} \vec{B_{ext}}.\vec{u_z} \, dz + \int_{bord~int} \vec{B_{int}}.\vec{u_z} \, dz $$

En effet, les intégrales le long des côtés selon la direction x sont nulles, car on a vu dans la question 1 que le champ magnétique était orienté selon la direction z. Les produits scalaires entre champ magnétique et vecteurs tangents sont donc nuls le long de ces côtés.

De plus, le champ à l'extérieur du conducteur ne pourrait être dirigé que selon la direction z vu la géométrie du problème, et on sait que $B_z$ est nul à l'extérieur (dit dans l'énoncé). Par conséquent le champ magnétique extérieur est nul, et la première intégrale de la somme précédente disparaît. On a finalement:

$$\oint_{\partial S} \vec{B}.\vec{dl}= \int_{bord~int} \vec{B_{int}}.\vec{u_z} \, dz=\int_{bord~int} \ B(x,y;t) \, \vec{u_z}.\vec{u_z} \, dz=\int_{bord~int} \ B_z \, dz$$

Vu que le champ magnétique est indépendant de z, on a alors:

$$\oint_{\partial S} \vec{B}.\vec{dl}=B_z\int_{bord~int} \ dz=B_z \, \Delta z  $$

On s'attarde maintenant sur le second membre de l'équation relative au théorème d'Ampère. On peut rappeler rapidement comment on l'obtient.
On commence par appliquer le théorème de Stokes au membre de gauche:

$$ \oint_{\partial S} \vec{B}.\vec{dl}=\iint_{S} \ 
(\vec{\nabla} \times \vec{B}).d\vec{S}$$

Et en utilisant une des équations de Maxwell, cette intégrale devient:

$$\iint_{S} \ 
\mu_0 \, \vec{J}.d\vec{S}=\mu_0\iint_{S} \ \, \vec{J}.d\vec{S}=\mu_0 \,I_{traversant} $$

On cherche à déterminer $I_{traversant}=\iint_{S} \ \, \vec{J}.d\vec{S}$ dans le cas de notre surface rectangulaire.
Pour faciliter le calcul du courant, on place l'origine d'un repère xz au milieu de notre rectangle. On voit alors que le rectangle est traversé (perpendiculairement) par le courant $\vec{J}$ (valant $1 \, A/m^2$) en chaque point le long du segment où x=0. En d'autres termes, dans le rectangle considéré, le courant $\vec{J}$ est tel que $\vec{J}=1 \,\vec{u_y}$ si x=0 et $\vec{J}=\vec{0}$ sinon. On peut traduire cela en écrivant $\vec{J}=1 \,\vec{u_y} \, \delta(x)$.

On prend comme vecteur unitaire normal à la surface $-\vec{u_y}$ (on fait ce choix car il va nous permettre d'obtenir un $B_z$ négatif), on peut alors écrire $d\vec{S}=-dx \, dz \, \vec{u_y}$. On sait maintenant calculer le courant la traversant:

$$ I_{traversant}=\iint_{S} \ \, \vec{J}.d\vec{S} =- \iint \ \vec{J}.\vec{u_y} \,dx \,dz=-\iint \ \vec{u_y} \, \delta(x) .\vec{u_y} \,dx \,dz=-\iint \ \delta(x) \,dx \,dz$$

Or, on sait que $\int \ \delta(x) \, dx=1$. On obtient finalement:

$$I_{traversant}=-\int dz \int \delta(x) \, dx=-\int dz=-\Delta z$$

En égalant nos résulats obtenus pour chaque membre de l'équation du théorème d'Ampère, on obtient:
$$B_z \, \Delta z=-\mu_0\Delta z$$
c'est-à-dire
$$B_z=-\mu_0$$

En répétant cette démarche pour chacun des bords externes du conducteur, on trouve à chaque fois comme conditions aux bords $B_z=-\mu_0$.

##### Bords du conducteur en contact avec le supraconducteur



On va maintenant chercher les conditions aux bords internes du conducteur.
On connaît la condition aux bords pour le champ électrique; sa composante tangentielle doit être continue:

$$\vec{n_{sc}}\times (\vec{E_{c}}-\vec{E_{s}})=\vec{0}$$

où $\vec{E_{c}}$ et $\vec{E_{s}}$ sont respectivement les champs électriques dans le conducteur et dans le supraconducteur, et où $\vec{n_{sc}}$ est le vecteur unitaire normal à l'interface, dirigé du supraconducteur vers le conducteur.
Or, on sait que le champ électrique dans le supraconducteur est toujours nul. En effet, on sait que la loi d'Ohm s'applique:

$$\vec{J}=\sigma\vec{E}$$
$$\vec{E}=\frac{\vec{J}}{\sigma}$$

La conductivité électrique $\sigma$ étant infinie dans un supraconducteur et la densité de courant étant finie pour avoir un sens physique, le champ électrique y est donc bien nul.

Notre condition aux bords pour le champ électrique devient donc:

$$\vec{n_{sc}}\times \vec{E_{c}}=\vec{0}$$
    
En utilisant d'abord la loi d'Ohm évoquée juste avant, puis l'équation de Maxwell $ \vec{\nabla} \times \vec{B}  = \mu_0 \vec{J}$, la condition devient:
$$\vec{n_{sc}}\times \vec{E_{c}}=\vec{n_{sc}} \times \frac{\vec{J}}{\sigma}=\vec{n_{sc}} \times \frac{(\nabla \times \vec{B})}{\sigma\mu_0}=\vec{0}$$

où $\vec{B}$ est le champ à l'intérieur du conducteur et où $\sigma$ est la conductivité électrique du conducteur.
Le facteur multiplicatif n'ayant pas d'influence, on obtient la condition

$$\vec{n_{sc}} \times (\nabla \times \vec{B})=\vec{0}$$

On a vu dans la question 1 que le champ magnétique est dirigé selon l'axe z: $\vec{B}=B_z\vec{u_z}$ avec $B_z$ négatif ou nul.
Par conséquent, le rotationnel de $\vec{B}$ est donné par:

$$\begin{vmatrix}\vec{u_x} & \vec{u_y} & \vec{u_z}\\\partial_x &\partial_y & \partial_z\\0 & 0 & B_z\end{vmatrix}=\partial_yB_z \,\vec{u_x}-\partial_xB_z\,\vec{u_y}$$

La condition devient finalement:

$$\vec{n_{sc}} \times (\partial_yB_z\vec{u_x}-\partial_xB_z\vec{u_y})=\vec{0}$$

Prenons le bord interne gauche. Sur ce bord, le vecteur $\vec{n_{sc}}$ est donné par $-\vec{u_x}$. La condition est donc 

$$-\vec{u_x} \times (\partial_yB_z\vec{u_x}-\partial_xB_z\vec{u_y})=\vec{u_x}\times\vec{u_x}\partial_xB_z=\partial_xB_z\vec{u_z}=\vec{0}$$

c'est-à-dire $\partial_xB_z=0$.

Prenons le bord interne supérieur. Sur ce bord, le vecteur $\vec{n_{sc}}$ est cette fois-ci donné par $\vec{u_y}$. La condition est donc

$$\vec{u_y} \times (\partial_yB_z\vec{u_x}-\partial_xB_z\vec{u_y})=\vec{u_y}\times\vec{u_x}\partial_yB_z=-\partial_yB_z\vec{u_z}=\vec{0}$$

c'est-à dire $\partial_yB_z=0$.

De la même manière, la condition sur le bord interne droit est $\partial_xB_z=0$ et celle sur le bord interne inférieur est $\partial_yB_z=0$.

Les conditions sur les bords du conducteur en contact avec le supraconducteur sont donc des conditions de Neumann.

##### Discrétisation de l'équation

On discrétise notre domaine en construisant une grille discrète telle que, dans la direction $x$, la discrétisation utilise les indices $i=0,1,...,n_x$ et dans la direction $y$, les indices $j=0,1,...,n_y$. Si le domaine a une longueur $L$ selon $x$ et $l$ selon $y$, l'écart entre deux points de la grille sera de $dx=\frac{L}{n_x-1}$ selon la direction $x$ et $dy=\frac{l}{n_y-1}$. Ainsi, si on s'intéresse à la valeur du champ $B$ en un point $(x=idx,y=jdy)$ de la grille, elle sera donnée par

$$B_{i,j}$$

Ensuite, on discrétise également le temps avec les indices $n=0,1,...,n_t$. Si le step de temps est $dt$, la valeur du champ $B$ en un point $(x=idx,y=jdy)$ de la grille en un temps $t=n dt$ est donnée par 

$$B_{i,j}^n$$

*Remarque* : Dans le but d'avoir une orientation de grille qui est naturelle vis-à-vis d'un repère cartésien, la donnée $B_{i,j}$ sera stockée dans le tableau avec les indices B[j,i]. Ainsi, l'indice de ligne correspond à $y$ et l'indice de colonne correspond à $x$.

Pour rappel, en notant notre coefficient de diffusion $\alpha$, l'équation de diffusion à laquelle on s'intéresse est :
$$\frac{\partial B}{\partial t} = \alpha \left(\frac{\partial^2 B}{\partial x^2} + \frac{\partial^2 B}{\partial y^2} \right)$$

Que l'on discrétise selon le schéma *forward-time/central-space* :

$$\frac{B^{n+1}_{i,j} - B^n_{i,j}}{\Delta t} = \alpha \left( \frac{B^n_{i+1, j} - 2B^n_{i,j} + B^n_{i-1,j}}{\Delta x^2} + \frac{B^n_{i, j+1} - 2B^n_{i,j} + B^n_{i,j-1}}{\Delta y^2}\right)$$

Pour résoudre le problème, l'étape à effectuer pour avancer dans le temps est  

$$B^{n+1}_{i,j}= B^n_{i,j} + \alpha \left( \frac{\Delta t}{\Delta x^2} (B^n_{i+1, j} - 2B^n_{i,j} + B^n_{i-1,j}) + \frac{\Delta t}{\Delta y^2} (B^n_{i, j+1} - 2B^n_{i,j} + B^n_{i,j-1})\right)$$

Dans ce problème, $\Delta x = \Delta y \equiv \delta$ donc on peut réécrire proprement :

$$B^{n+1}_{i,j}= B^n_{i,j} + \alpha \frac{\Delta t}{\delta^2} \left(( B^n_{i+1, j} - 2B^n_{i,j} + B^n_{i-1,j}) +( B^n_{i, j+1} - 2B^n_{i,j} + B^n_{i,j-1})\right)$$

C'est un schéma explicite car il utilise les données de l'étape précédente pour avancer dans le temps.

##### Condition de Neumann

La grille spatiale que nous devons considérer doit comporter 121x121 points. On sait qu'un côté du supraconducteur mesure 0,5cm et qu'un coté du conducteur mesure 1,5cm. On en déduit rapidement que ce sont les 40ème et 80ème points de la grille (en les comptant à partir de 0) selon un axe ou l'autre qui vont nous permettre de délimiter le supraconducteur.

Reprenons comme exemple notre condition pour le bord interne gauche du conducteur. Celle-ci est donnée par $\partial_xB_z=0$. Cette équation signifie que la variation du champ magnétique le long de l'axe x en restant dans le conducteur est nulle au voisinage direct de ce bord. En terme de la discrétisation choisie, cela se traduit de la façon suivante:

$$ B^n_{40,~j}=B^n_{39,~j}~~~~\forall n,~~~40\le j \le 80 $$

De la même manière, pour le bord interne supérieur, la condition est donnée par $\partial_yB_z=0$: la variation du champ magnétique le long de l'axe y en restant dans le conducteur est nulle au voisinage direct du bord. En terme de discrétisation, on a:

$$ B^n_{i,80}=B^n_{i,81}~~~~\forall n,~~~40\le i \le 80 $$

La démarche est identique pour les 2 bords internes restants. Pour le bord de droite, on obtient:

$$ B^n_{80,~j}=B^n_{81,~j}~~~~\forall n,~~~40\le j \le 80$$
Et pour le bord inférieur:
$$ B^n_{i,40}=B^n_{i,39}~~~~\forall n,~~~40\le i \le 80$$

##### Stabilité

On a vu que pour une équation de diffusion 1D en schéma explicite reste stable si $\alpha \frac{\Delta t}{(\Delta x)^2} \leq \frac{1}{2}$. En 2D, il faut étendre ça dans les deux directions, c'est-à-dire : 

$$\alpha \frac{\Delta t}{(\Delta x)^2} + \alpha \frac{\Delta t}{(\Delta y)^2} \leq \frac{1}{2}$$

Comme  $\Delta x = \Delta y \equiv \delta$, il faut 

$$\alpha \frac{\Delta t}{\delta^2} \leq \frac{1}{4}$$

##### Résolution

Dans un premier temps, importons les librairies requises et définissons les constantes et grandeurs utiles du problème.

```python
import numpy
from matplotlib import pyplot
%matplotlib inline
from matplotlib import rcParams, cm
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
from matplotlib import animation
from IPython.display import HTML
```

```python
## Constantes 

sigma=5.96*10**7    #S/m
mu_0=1.26*10**(-6)  #N/A²
alpha=1/(sigma*mu_0)
```

```python
## Conditions

nx=121
ny=121

L=0.015            #m #longueur coté carré

dx=L/(nx-1)
dy=L/(ny-1)

dt=2.5*10**(-7)    #seconde
nt=5000            #pour arriver au temps désiré pour les questions 1 à 6

x=numpy.linspace(0,L,nx)         #utiles pour le futur graphique
y=numpy.linspace(0,L,ny)
```

Vérifions la condition de stabilité

```python
print(alpha*dt/dx**2)
```

C'est inférieur à 0.25 : la valeur de $dt$ choisie est donc bien justifiée.

Imposons maintenant que le champ soit initialement nul partout sur le domaine et qu'il valle $-\mu_0$ (Dirichlet) sur les bords extérieurs. 

```python
Bi = numpy.zeros((ny, nx))
Bi[0,:]= - mu_0
Bi[:,0] = - mu_0
Bi[:,nx-1]= - mu_0    #shift de 1 dû à la numérotation de Python
Bi[ny-1,:]= - mu_0
```

Définissons la fonction qui effectuera la diffusion.

Au sein de celle-ci, il faudra imposer $B=0$ dans le supraconducteur à toutes les itérations dans le temps.

*Pourquoi ?*

On a vu précédemment que le champ électrique était nul dans le supraconducteur. Or, en vertu de l'équation de Maxwell $\vec{\nabla} \times \vec{E} = -\partial_t \vec{B}$, le champ magnétique est donc une constante du temps dans le supraconducteur. Il reste donc toujours nul vu que $B=0$ dans tout le domaine initialement.

```python
def diffusion(B,nt,alpha,dt,dx,dy):
    for n in range(nt):
        Bn = B.copy()
        B[1:-1,1:-1] = Bn[1:-1,1:-1] + alpha *\
            (dt/dy**2 * (Bn[2:,1:-1] - 2*Bn[1:-1,1:-1] + Bn[:-2,1:-1]) +\
             dt/dx**2 * (Bn[1:-1,2:] - 2*Bn[1:-1,1:-1] + Bn[1:-1,:-2]))
        
        
        ## Forçage de champ nul dans le carré central
        B[41:80,41:80]=0
        
        ##Conditions de Neumann carré central
            #Coté droit du carré
        B[40:81,80]=B[40:81,81]
            #Coté haut
        B[80,40:81]=B[81,40:81]
            #Coté gauche
        B[40:81,40]=B[40:81,39]
            #Coté bas
        B[40,40:81]=B[39,40:81]
        
        
                
    return B
```

Intéressons-nous donc à l'évolution de $B$ jusqu'au temps $t=1,25 . 10^{-3} s$.

```python
B=diffusion(Bi.copy(),nt,alpha,dt,dx,dy)
```

```python
pyplot.figure(figsize=(10,8))
pyplot.title('Champ magnétique au temps $t=1,25 . 10^{-3} s$',fontsize=20)
pyplot.contourf(x,y,B,150,cmap=cm.viridis)
pyplot.xlabel('$x \,(m)$')
pyplot.ylabel('$y \,(m)$')
pyplot.colorbar();
```

Le champ minimum se diffuse bien des bords extérieur vers l'intérieur du domaine tandis que le "carré" supraconducteur reste avec un champ nul et il n'y a pas de flux du champ entrant ou sortant de ce supraconducteur. On s'attend à ce que si l'on fait tendre le temps vers l'infini (ou au moins vers une valeur très grande), on se retrouve avec un champ uniforme et mininum partout sur le domaine sauf en le carré central dans lequel il sera toujours nul. On remarque cependant une fine couche autour du carré supraconducteur dans laquelle le champ a une valeur intermédiaire (en bleu ciel) ; ceci est dû à la discrétisation et au fait que l'évolution est réalisée sur un schéma centré d'espace.

*Remarque 1* : Les valeurs de $B$ sont en Tesla.

*Remarque 2* : Le problème étant indépendant de $z$, cela peut donc correspondre à une coupe dans le matériau en $z=0$ par exemple.

### Question 4

Pour l'animation, on va de nouveau faire évoluer le problème grâce à la même boucle que précédemment mais en stockant, à chaque itération, une image (grâce à *imshow*) dans une grande liste d'images. Créons une fonction qui génère l'animation pour pouvoir modifier différents paramètre à notre guise. Il est à remarquer qu'à plusieurs endroits, nous avons besoin de "forcer" des divisions à être entières dans le but d'utiliser leur résultat comme indice ou comme limite de boucle. Ceci en fait n'est rien d'autre qu'une formalité au niveau du code mais ça ne change rien car les divisions de base donnaient quand même un résultat entier.

```python
def animation_maker(t,images,ti,B,alpha,dt,dx,dy):
    
    #t est le temps jusqu'où va aller l'animation (en secondes)
    #images est le nombre d'images de l'animation
    #ti est le temps d'affichage de chaque image dans l'animation (en millisecondes)
    
    fig=pyplot.figure(figsize=(10,8))
    pyplot.title('Évolution du champ magnétique',fontsize=20)
    pyplot.xlabel('$x \,(m)$')
    pyplot.ylabel('$y \,(m)$')
    
    ims=[j for j in range(images)]        #initialisation de la liste d'images
    
    nt=int(t/dt)                          #nombre de time steps requises
    
    interv=int(nt/images)                 #nombre d'itérations entre deux rajouts d'images dans ims
                                          
    
    for n in range(nt):
        Bn = B.copy()
        B[1:-1,1:-1] = Bn[1:-1,1:-1] + alpha *\
            (dt/dy**2 * (Bn[2:,1:-1] - 2*Bn[1:-1,1:-1] + Bn[:-2,1:-1]) +\
             dt/dx**2 * (Bn[1:-1,2:] - 2*Bn[1:-1,1:-1] + Bn[1:-1,:-2]))
        
        ## Forçage de champ nul dans le carré central
        B[41:80,41:80]=0
        
        ##Conditions de Neumann carré central
            #Coté droit du carré
        B[40:81,80]=B[40:81,81]
            #Coté haut
        B[80,40:81]=B[81,40:81]
            #Coté gauche
        B[40:81,40]=B[40:81,39]
            #Coté bas
        B[40,40:81]=B[39,40:81]       
      
        ##On ne rajoute qu'une image toutes les 'interv' images
        if(n%interv ==0):
            im = pyplot.imshow(B, animated=True,extent=[0,L,0,L])
            ims[int(n/interv)]=[im]
   
    return animation.ArtistAnimation(fig, ims, interval=ti, blit=True)
```

Ce qui donnera pour $t=1,25 . 10^{-3} s$ et avec 100 images : 

```python
HTML(animation_maker(1.25*10**(-3),100,75,Bi.copy(),alpha,dt,dx,dy).to_html5_video())
```

Nous nous pouvons malheureusement pas empêcher l'affichage (en plus de la vidéo) de la dernière image rangée dans *ims* lorsque l'on exécute la fonction.

### Question 6 

Pour rappel, l'équation de Maxwell reliant le rotationnel de $\vec{B}$ et $\vec{J}$ permet d'obtenir : 

$$ \frac{\partial B_z(x,y)}{\partial y} = \mu_0 \, J_x(x,y)$$
$$ \frac{\partial B_z(x,y)}{\partial x} = - \mu_0 \, J_y(x,y)$$

Et on peut donc discrétiser les dérivées (selon une méthode centrée d'espace) : 

$$ \frac{B_z(x,y+\Delta y) -B_z(x,y-\Delta y)}{2\Delta y} = \mu_0 \, J_x(x,y)$$
$$ \frac{B_z(x+\Delta x,y) -B_z(x-\Delta x,y)}{2\Delta x} = -\mu_0 \, J_y(x,y)$$

ce qui nous donne facilement accès aux composantes $J_x$ et $J_y$ de ce courant surfacique (la composante selon $z$ est nulle vu que $\vec{B}$ est exclusivement selon $z$).

```python
def courant(x,y):
    i=int(x/dx)
    j=int(y/dy)
    Jx=(B[j+1,i]-B[j-1,i])/(2*dy*mu_0)
    Jy=(B[j,i-1]-B[j,i+1])/(2*dx*mu_0)
    return Jx, Jy
```

```python
courant(2.5*10**(-3),2.5*10**(-3))
```

Le courant électrique surfacique à l'instant $t=1,25 . 10^{-3} s$ en $(2,5 . 10^{-3} m ; 2,5 . 10^{-3} m)$ vaut donc 

$$ \vec{J} = (\, 22.1300479 \,; \, -22.1300479 \, ; \, 0\,) \, A/m^2 $$

Ce qui veut dire qu'à cet endroit du conducteur, le courant "tourne" en fait dans le sens opposé au courant appliqué sur la surface.

### Question 7 

On aimerait maintenant déterminer la valeur du flux du champ magnétique à travers la section z=0 par exemple, à un temps donné. Pour rappel, le flux $\phi$ est donné par:

$$\phi=\iint_{S} \ \, \vec{B}.d\vec{S}$$

Vu qu'on travaille en discrétisant notre surface, on traduit cette expression du flux dans le cas continu à une expression dans le cas discret:

$$\phi=\sum_{i}\vec{B_i} .\vec{\Delta S_i}$$

où $\vec{B_i}$ est le champ magnétique traversant la surface $\Delta S_i$ et où $\vec{\Delta S_i}$ est normal à la surface $\Delta S_i$.

Dans notre cas, chaque $\vec{B_i}$ correspondra à la valeur du champ en un point de la grille spatiale et chaque $\Delta S_i$ correspondra au petit carré de côté dx dont le milieu est le point de la grille spatiale en question. De plus, on choisit de prendre $\vec{\Delta S_i}=dx\,dy\, \vec{u_z}=\delta^2 \vec{u_z}$. Le champ magnétique étant toujours dirigé dans le même sens ($B_z$ est toujours négatif ou nul), tous les éléments de flux qu'il faudra sommer pour obtenir le flux total seront négatifs.

On remarque que si on emploie la méthode évoquée juste avant, c'est-à-dire que si on encadre chaque point de la grille d'un petit carré de côté dx, la surface comptée sera trop grande. En effet, en encadrant de carrés les points des bords externes du conducteur, on va "dépasser" du conducteur et empiéter sur l'extérieur. Cependant, dans la limite d'une grille spatiale constituée d'une infinité de points, la contribution au flux apportée par les éléments du bord tendra vers 0. Le choix de la gestion des éléments du bord n'est donc pas unique et mène à la même valeur de flux pour une grille dont le nombre d'éléments tend vers l'infini.
Nous avons personnellement choisi de ne pas utiliser les valeurs de champ magnétique aux points situés sur les bords de la grille spatiale et donc de calculer le flux sur une surface un peu plus petite que celle de la section du conducteur.

Définissons donc une fonction qui permet de connaitre le flux de notre champ magnétique à travers la section à n'importe quel moment. Bien entendu, pour cela, il faut recommencer une diffusion de notre champ $B$. On ne va pas faire dépendre cette fonction des paramètres initiaux de la diffusion ($B_i,\alpha,dt,...$) car on suppose que l'on s'intéressera toujours au même problème.

```python
def flux(t):
    B=diffusion(Bi.copy(),int(t/dt),alpha,dt,dx,dy)
    somme=0
    for i in range(1,121):
        for j in range(1,121):
            somme += B[i,j]
    return somme*dx*dy
```

```python
t=1.25*10**(-2)
flux(t)
```

Ainsi, le flux qui traverse la section du matériau au temps $t=1,25 . 10^{-2} s$ est

$$ \phi = -2,53555312 . 10^{-10} \, Tm^2 $$
