```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
T=100.0
dt=0.02
N=int(T/dt)+1
t=numpy.linspace(0.0,T,N) #Start,Stop,Step
```

Méthode de Euler

```python
#conditions initiales
z0=100.
b0=10.
zt=100.
g=9.81

u=numpy.array([z0,b0])
print(u)
```

```python
#initialisation d'un vecteur pour suivre le changement des z
z=numpy.zeros(N)
z[0]=z0
```

```python
#résolution avec Euler
for n in range(1,N):
    u = u + dt*numpy.array([u[1],g*(1-u[0]/zt)]) 
    z[n]=u[0]
#on fait évoluer les deux cases du vecteur en même temps
#on garde une trace de l'évolution de z dans notre tableau
#rempli de 0
```

```python
pyplot.figure(figsize=(10,6))  #taille du graph
pyplot.ylim(50,150)            #limites axe y 
pyplot.tick_params(axis='both',labelsize=14) #taille police
pyplot.xlabel('t',fontsize=14) #nom axe x
pyplot.ylabel('z',fontsize=14) #nom axe y
pyplot.plot(t,z,'k-')          #création graph
```

Solution exacte

```python
z_exact=b0*(zt/g)**.5*numpy.sin((g/zt)**.5*t)+\
(z0-zt)*numpy.cos((g/zt)**.5*t)+zt
```

```python
pyplot.figure(figsize=(10,6))  #taille du graph
pyplot.ylim(40,160)            #limites axe y 
pyplot.tick_params(axis='both',labelsize=14) #taille police
pyplot.xlabel('temps',fontsize=14) #nom axe x
pyplot.ylabel('z',fontsize=14) #nom axe y
pyplot.plot(t,z,)              #création graph1
pyplot.plot(t,z_exact)         #création graph2
pyplot.legend(['Solution numérique','Solution analytique'])
```
